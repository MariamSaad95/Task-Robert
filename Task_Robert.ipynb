```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imshow,imread,imsave
import cv2
```

```python
img=cv2.imread('edge.jpg')
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
img=cv2.GaussianBlur(img,(3,3),0)
```

```python
Robert_x=np.array([[1,0],[0,-1]])
Robert_y=np.array([[0,1],[-1,0]])
```

```python
img_Robert_x=cv2.filter2D(img,-1,Robert_x)
img_Robert_y=cv2.filter2D(img,-1,Robert_y)
```

```python
edged_img = np.sqrt( np.square(img_Robert_x) + np.square(img_Robert_y))
```

```python
plt.imshow(img_Robert_x,cmap='gray')
plt.title('img_Robert_x')
plt.show()

plt.imshow(img_Robert_y,cmap='gray')
plt.title('img_Robert_y')
plt.show()

plt.imshow(edged_img,cmap='gray')
plt.title('edged_img')
plt.show()
```
